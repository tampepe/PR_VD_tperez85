```python
import pandas as pd
```

```python
# Lista de años a procesar
years = [2017, 2018, 2019, 2020, 2021]
all_dfs = []

# Cargar y procesar cada archivo desde el subdirectorio "Hours worked/"
for year in years:
    file_path = f"Parte1/Hours worked/hours_worked_{year}.csv"
    
    df = pd.read_csv(
        file_path,
        sep=';',
        skiprows=3,
        names=['country', 'hours_worked'],
        decimal=','
    )
    df['year'] = year
    all_dfs.append(df)
```

```python
# Unir todos los dataframes
combined_df = pd.concat(all_dfs, ignore_index=True)

# Agrupar por país y calcular la media de horas trabajadas
hours_worked_df = (
    combined_df
    .groupby('country')['hours_worked']
    .mean()
    .reset_index(name='avg_hours_worked')
)

hours_worked_df.head()
```

```python
# Leer el archivo Excel, hoja principal
happiness_df = pd.read_excel(
    "Parte1/Happiness/Data for Figure 2.1 (2011–2024).xlsx", 
    sheet_name='Data for Figure 2.1 (2011–2024)' 
)
happiness_df.head()
```

```python
# Filtrar solo los años 2017 a 2021
happiness_filtered = happiness_df[happiness_df['Year'].between(2017, 2021)]

# Eliminar la columna 'Year' para que no entre en el cálculo de la media
happiness_filtered = happiness_filtered.drop(columns=['Year'])

# Agrupar por país y calcular la media de las demás columnas numéricas
happiness_df = (
    happiness_filtered
    .groupby('Country name')
    .mean(numeric_only=True)
    .reset_index()
    .rename(columns={'Country name': 'country'})
)

happiness_df.head()
```

```python
def clasificar_gasto(row):
    if row < 5:
        return 'Bajo'
    elif row < 8:
        return 'Medio'
    else:
        return 'Alto'

merged_df['health_exp_category'] = merged_df['avg_health_exp'].apply(clasificar_gasto)
```

```python
# 4. Clasificar según los rangos oficiales del UNDP
def clasificar_hdi(hdi):
    if hdi < 0.550:
        return 'Bajo'
    elif hdi < 0.700:
        return 'Medio'
    elif hdi < 0.800:
        return 'Alto'
    else:
        return 'Muy alto'
```

```python
# Corregir ranking columna happiness_rank_country
ranking_df = (
    final_df.groupby("country")["happiness_rank_country"]
    .agg(lambda x: x.mode().iloc[0]) 
    .astype(int)                      
    .reset_index()
)

final_df = final_df.drop(columns=["happiness_rank_country"]).merge(ranking_df, on="country", how="left")
final_df["happiness_rank_country"].unique()
```

```python
# Normalizar nombres de países para Flourish
country_mapping = {
    "United States": "United States of America",
    "United Kingdom": "United Kingdom",
    "Macedonia": "North Macedonia"
}

final_df["country"] = final_df["country"].replace(country_mapping)
```

```python
# Generar CSV para Flourish con variables estructurales normalizadas entre 0 y 1
# Lista de columnas seleccionadas
columns = [
    "avg_health_exp_country",
    "avg_hdi_country",
    "happiness_rank_country", # está columna será invertida (ranking 1 = país más feliz)
    "life_satisfaction_score_country",
    "gdp_log_country",
    "social_support_country",
    "healthy_life_expectancy_country",
    "freedom_life_choices_country",
    "generosity_country",
    "corruption_perception_country"
]

# Agrupar por país
map_df = final_df.groupby("country")[columns].mean().reset_index()

# Invertimos el ranking para que "más feliz" (puesto 1) tenga valor alto al normalizar
map_df["happiness_rank_country"] = -map_df["happiness_rank_country"]

# Normalización manual (min-max) con Pandas
for col in columns:
    min_val = map_df[col].min()
    max_val = map_df[col].max()
    map_df[col] = (map_df[col] - min_val) / (max_val - min_val)

# Exportar CSV final
map_df.to_csv("csv/flourish_map_structural_variables_normalized.csv", index=False)
```

```python
# Flourish: Empresas tecnológicas vs no tecnológicas
# Variables que vamos a agrupar
cols_to_analyze = ["benefits", "workplace_resources", "medical_coverage"]

# Inicializamos un DataFrame vacío
grouped_data = []

# Agrupamos por tipo de empresa y calculamos el % de "Yes" para cada variable
for col in cols_to_analyze:
    temp = (
        final_df.groupby("tech_company")[col]
        .value_counts(normalize=True)  # porcentaje
        .rename("percentage")
        .reset_index()
    )
    # Filtramos solo las respuestas "Yes"
    temp = temp[temp[col] == "Yes"]
    temp["variable"] = col
    grouped_data.append(temp[["tech_company", "variable", "percentage"]])

# Unimos todos los resultados
bar_df = pd.concat(grouped_data)

# Pivotamos para que cada variable sea una columna (opcional para Flourish)
bar_pivot = bar_df.pivot(index="tech_company", columns="variable", values="percentage").reset_index()

# Exportar CSV
bar_pivot.to_csv("csv/flourish_bar_tech_vs_support.csv", index=False)
```

```python
# Calcular el porcentaje de personas con un trastorno que pueden hablar con su jefe
# y cruzarlo con el nivel de libertad de vida en su país

# Filtrar personas que declaran tener un trastorno de salud mental
df_filtered = final_df[final_df["mental_health"] == "Yes"].copy()

# Limpiar posibles espacios en las respuestas
df_filtered["mh_employer_discussion"] = df_filtered["mh_employer_discussion"].str.strip()

# Filtrar respuestas válidas en la pregunta sobre apertura con el empleador
valid = df_filtered[df_filtered["mh_employer_discussion"].isin(["Yes", "No", "Maybe"])]

# Calcular el total de personas con trastorno por país
total = valid.groupby("country").size().rename("total_with_disorder")

# Calcular cuántas de esas personas respondieron "Yes"
yes = valid[valid["mh_employer_discussion"] == "Yes"].groupby("country").size().rename("yes_discuss")

# Calcular el porcentaje de apertura
summary = pd.concat([total, yes], axis=1).fillna(0)
summary["percent_discuss_with_disorder"] = summary["yes_discuss"] / summary["total_with_disorder"]

# Añadir la variable estructural 'freedom_life_choices_country' por país
freedom = final_df[["country", "freedom_life_choices_country"]].drop_duplicates()

# Unir los dos conjuntos de datos
plot_df = summary.reset_index().merge(freedom, on="country")

# Exportar el conjunto final a un archivo CSV para usar en Flourish
plot_df.to_csv("csv/flourish_scatter_freedom_vs_discussion_with_disorder.csv", index=False)
```

```python
# Filtrar personas que declaran tener un trastorno
filtered_df = final_df[final_df["mental_health"] == "Yes"].copy()
filtered_df["mh_employer_discussion"] = filtered_df["mh_employer_discussion"].str.strip()

# Filtrar respuestas válidas
valid = filtered_df[filtered_df["mh_employer_discussion"].isin(["Yes", "No", "Maybe"])]

# Calcular apertura por país
total_with_disorder = valid.groupby("country").size().rename("total_with_disorder")
yes_discuss = valid[valid["mh_employer_discussion"] == "Yes"].groupby("country").size().rename("yes_discuss")

# Calcular porcentaje
discussion_df = pd.concat([total_with_disorder, yes_discuss], axis=1).fillna(0)
discussion_df["percent_discuss_with_disorder"] = discussion_df["yes_discuss"] / discussion_df["total_with_disorder"]
discussion_df = discussion_df.reset_index()

# Calcular % "Yes" por país en variables de apoyo
support_vars = ["benefits", "workplace_resources", "medical_coverage"]
support_data = []

for var in support_vars:
    temp = (
        final_df[final_df[var].isin(["Yes", "No"])]
        .groupby("country")[var]
        .value_counts(normalize=True)
        .rename("percentage")
        .reset_index()
    )
    temp = temp[temp[var] == "Yes"]
    temp = temp[["country", "percentage"]].rename(columns={"percentage": f"percent_{var}"})
    support_data.append(temp)

support_df = support_data[0]
for other in support_data[1:]:
    support_df = support_df.merge(other, on="country", how="outer")  # usar outer aquí para no perder países útiles

# Añadir variables estructurales
structural_vars = [
    "avg_hdi_country",
    "freedom_life_choices_country",
    "life_satisfaction_score_country",
    "corruption_perception_country",
    "avg_health_exp_country",
    "social_support_country",
    "healthy_life_expectancy_country",
    "gdp_log_country",
    "generosity_country"
]
structural_df = final_df[["country"] + structural_vars].drop_duplicates()

# Combinar
merged = discussion_df.merge(support_df, on="country", how="left")
merged = merged.merge(structural_df, on="country", how="left")

# Calcular tamaño del punto (para scatter plots) y marcar si tiene muestra suficiente
merged["point_size"] = np.sqrt(merged["total_with_disorder"].fillna(0))
merged["sufficient_sample"] = merged["total_with_disorder"] >= 5

# Exportar CSV final
merged.to_csv("csv/df_scatterplot_general_flexible.csv", index=False)
```
